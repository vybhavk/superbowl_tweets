```python
def sentiment_ratio(keyword,twindow):
    df_window = df[[twindow[0]<time<twindow[1] for time in df.hour_offset]]
    thresh = 0.2
    df_keyword = df_window[[keyword in entry for entry in df_window.content]]
    return df_keyword[df_keyword.polarity>thresh].polarity.count() / df_keyword[df_keyword.polarity<-thresh].polarity.count().astype('float')
```

```python
def keyword_count(keyword,twindow):
    df_window = df[[twindow[0]<time<twindow[1] for time in df.hour_offset]]
    df_keyword = df_window[[keyword in entry for entry in df_window.content]]
    count = df_keyword.content.count()
    return count
```

```python
counts = []
for i in df_ads[['keyword','window']].values:
    count = keyword_count(i[0],i[1])
    counts.append(count)
df_ads['count'] = counts
```
